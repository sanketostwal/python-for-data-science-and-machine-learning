**Matplotlib Part 1**

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2

print (x)
print (y)
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ABCD')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.subplot
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
plt.plot(x,y)
```

**Matplotlib part 2**

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
axes[-1,1].plot(x,y)
plt.tight_layout()
```

**Figure Size and DPI**

```python
fig,axes = plt.subplots(ncols=1,nrows=1,figsize=(5,5))

#axes[0].plot(x,y)
axes.plot(x*y,x+y)
```

```python
fig.savefig('test.jpg',dpi=200,axes=1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label ='Square Plot')
ax.plot(x,x**3,label ='Cube Plot')

ax.legend(loc=0)
```
